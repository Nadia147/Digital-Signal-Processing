```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
import math
```

```python
#constracting with sine cosine &dc
state =256
t=np.arange(0,1,1/state)
amp=3
f=5
x1=amp*np.sin(2*np.pi*f*t)
x2=4*np.sin(2*np.pi*3*t)
x=x1+x2+3
plt.figure(figsize=(24,10))


plt.title('',fontsize=25)
plt.subplot(2,2,3)
plt.plot(t,x,linewidth=3)

plt.title('Combine signal',fontsize=25)
plt.xlabel('Time in sec',fontsize=25)
plt.ylabel('amplatude',fontsize=25)
#lt.subplot(2,2,2)
#plt.plot(t,x3,linewidth=3)
plt.plot()
```

```python
state =256
t=np.arange(0,1,1/state)
amp=3
f=5
x1=amp*np.sin(2*np.pi*f*t)
x2=4*np.sin(2*np.pi*3*t)
x=x1+x2+3

#x=x4
X=np.zeros(len(x),dtype=complex)
for freq in range(0,len(t)) :
  #create cpx sin wv &dot product with sig
  csw=np.exp(-1j*2*np.pi*freq*t)
  X[freq]=np.sum(np.multiply(x,csw))

#extract amplitude
amps=np.abs(X)/len(t)
```

```python
plt.figure(figsize=(14,10))
plt.suptitle('signal and fouriar transfarm',fontsize=40)
plt.subplot(2,2,1)
plt.plot(t,x,linewidth=3)
plt.title('amp=3 freq=5')
plt.xlabel('Time in sec',fontsize=25)
plt.ylabel('amplatude',fontsize=25)

#plt.title('sin wv with apm 5 freq 2',fontsize=25)
plt.subplot(2,2,2)
markline,baseline,steanline=plt.stem(amps)
plt.setp(steanline,'linewidth',3)
plt.title('foriear transform of a signal',fontsize=25)
plt.xlabel('Indices',fontsize=25)
plt.ylabel('amplatude',fontsize=25)
plt.show()
```

```python
#correcting amp and converting indices to freq
state =256
t=np.arange(0,1,1/state)
amp=3
f=5
x1=amp*np.sin(2*np.pi*f*t)
x2=amp*np.cos(2*np.pi*3*t)
x=x1+x2+3
X=np.zeros(len(x),dtype=complex)
for freq in range(0,len(t)) :
  #create cpx sin wv &dot product with sig
  csw=np.exp(-1j*2*np.pi*freq*t)
  X[freq]=np.sum(np.multiply(x,csw))

#extract amplitude
amps=2*np.abs(X)/len(t)
#incorporate the neg freq of complex sinusoydal


```

```python
plt.figure(figsize=(14,8))
plt.title('Energy Spectram',fontsize=25)
amps=amps**2
plt.plot(t,amps,linewidth=3)
```

```python
nyquist = state/2
hz=np.linspace(0,nyquist,math.floor(len(t)/2+1))
```

```python
#plotting the signal and its fouriar transform
plt.figure(figsize=(24,9))
plt.suptitle('')
plt.subplot(2,2,1)
plt.plot(t,x,linewidth=3)
plt.title('amp=3 freq=5')
plt.xlabel('Time in sec',fontsize=25)
plt.ylabel('amplatude',fontsize=25)

#plt.title('sin wv with apm 5 freq 2',fontsize=25)
plt.subplot(2,2,2)
markline,baseline,steanline=plt.stem(hz,amps[range(0,len(hz))])
plt.setp(steanline,'linewidth',3)
plt.xlim(0,10)
plt.ylim(0,5)
plt.title('foriear transform of a signal',fontsize=25)
plt.xlabel('Freq',fontsize=25)
plt.ylabel('amplatude',fontsize=25)
plt.show()
```
