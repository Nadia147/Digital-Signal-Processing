```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#generating and ploting x2
x2=np.array([0,1,2,2,3,2,3])
n2=np.array([0,1,2,3,4,5,6])
plt.figure(figsize=(10,8))
markerline, stemlines, baseline = plt.stem(n2,x2, label='x2')
plt.setp(stemlines, 'linewidth', 1)
plt.legend(fontsize=30)
plt.show()

#shifting n+1

n3 = np.array([-1,0,1,2,3,4,5])
plt.figure(figsize=(10,8))

markerline, stemlines, baseline = plt.stem(n3,x2, label='shifting(n+1)')
plt.setp(stemlines, 'linewidth', 1)
plt.legend(fontsize=30)
plt.show()
```

```python
#generating and ploting x1
x1=np.array([1,2,2,1,1,3,5])
n1=np.array([0,1,2,3,4,5,6])
plt.figure(figsize=(20,10))
style.use('dark_background')
plt.subplot(2,2,1)
markerline, stemlines, baseline = plt.stem(n1,x1, label='x1')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)

#generating and ploting x2
x2=np.array([0,1,2,2,3,3,4])
n2=np.array([0,1,2,3,4,5,6])
plt.subplot(2,2,2)
markerline, stemlines, baseline = plt.stem(n2,x2, label='x2')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)

plt.subplot(2,2,3)
markerline, stemlines, baseline = plt.stem(n1,x1, label='x1')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)


#flip fold x2
f_x2 = x2[::-1]#slicing
n3 = np.array([-6,-5,-4,-3,-2,-1,0])
plt.subplot(2,2,4)
markerline, stemlines, baseline = plt.stem(n3,f_x2, label='flip/flod x2')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)
plt.show()
```

```python
#defile analog Signal
t = np.arange(0,11,1) 
x = np.sin(t)
```

```python
#signal denoising 

#create a signal
state = 255 #sampling rate 
t = np.arange(0,2,1/state)
points = len(t)
print(points)
print(t)
```

```python
#Creating original signal
x=np.sin(3*np.pi*t)
plt.figure(figsize=(20,10))
style.use('dark_background')

plt.plot(t,x, label='Original signal')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)
plt.show
```

```python
#Create a random noise
ns=4/np.random.randn(points)
plt.figure(figsize=(20,10))
style.use('dark_background')


# adding noise to orizinal signal
nssig = x+ns
plt.figure(figsize=(20,10))
style.use('dark_background')
plt.plot(t,nssig, label='Noisy signal')
plt.setp(stemlines, 'linewidth', 3)
plt.legend(fontsize=30)
plt.show
```

```python
#design filter for denoising

flt = 4.70*np.ones(50)/10.0


#filtering by convolution 
fltsig=np.convolve(nssig,flt,mode='full')
#ploting fit sig
plt.figure(figsize=(20,10))
plt.plot(fltsig, label='filtered Result')
plt.legend(fontsize=30)
plt.show()
```
