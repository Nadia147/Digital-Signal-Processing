Q.1). We are going to design a digital Radar target detection system. In radar system, there are basically two
signals, the first one is the sampled version of the transmitted signal sent to space and the second one is the sampled version of the received signal from space. If a target is present in the space being searched by radar, the
received signal consists of a delayed version of the transmitted signal with noise. Let, represent the received signal as follows: y(n) = ax(n-D) + w(kn).

x(n) = {3,2,1,1,2,3}; w(n) = {1.5,0.75,1.5}; Attenuation factor(a) = 2.0; Delay(D) = 1. Scale factor(k) = 2.

(i) Compute  y(n) step by step . Finally sketch.

(ii) Interpret the response of the system by applying signal correlation(Auto). Finally sketch.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# ploting x(n).
x=np.array([3,2,1,1,2,3,0])
n=np.array([-4,-3,-2,-1,0,1,2])
plt.figure(figsize=(20,10))
style.use('dark_background')
plt.subplot(3,1,1)
markerline,stemlines,baseline=plt.stem(n,x,label="x(n)")
plt.setp(stemlines,'linewidth',3)
plt.legend(fontsize=30)


```

```python
#plotting w(n)

plt.figure(figsize=(20,10))
style.use('dark_background')
w=np.array([0,0,1.5,0.75,1.5,0,0])
plt.subplot(3,2,1)
markerline,stemlines,baseline=plt.stem(n,w,label="w(n)")
plt.setp(stemlines,'linewidth',3)
plt.legend(fontsize=30)
```

```python
#plotting y(n)
plt.figure(figsize=(20,10))
style.use('dark_background')
xss=np.array([0,3,2,1,1,2,3]) 
xff=xss*2 
wss=np.array([0,0,0,1.5,1.5,0,0])
y=xff+wss

plt.subplot(3,1,2)
markerline,stemlines,baseline=plt.stem(n,y,label="Y(n)")
plt.setp(stemlines,'linewidth',3)
plt.legend(fontsize=30)
```

```python
# co-relation
r_sig=np.correlate(y,x,mode='same')

#r_sig /= r_sig[ 10- 1]


plt.figure(figsize=(20,10))
markerline,stemlines,baseline=plt.stem(n,r_sig,label="CO_RELATION")
plt.setp(stemlines,'linewidth',3)
plt.legend(fontsize=30)
plt.show()
```

ii)The following signal x(n) is transmitted. A random noise w(n) is added to it. Your task is to design a simple system to denoise the signal using convolution. The impulse(filter) for denoising is F(n). x(n) = 2.5(n+2) + 1.5(n+1) + 2.0(n) + 2.5(n-2); w(n) = {0.50, 1.0, 0.50, 0.80}; F(n) = {0.50, 0.50, 0.50} Finally sketch the denoised signal.

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#create a noiseless signl
x=np.array([2.5,1.5,2.0,0,2.5,0])
n=np.array([-2,-1,0,1,2,3])
#Create a random nose
ns=np.array([0,.5,1.0,.5,.8,0])
# adding noise signal
nssig = x+ns
```

```python
#ploting the original and noise signal
plt.figure(figsize=(30,10))
plt.subplot(2,1,1)
markerline,stemlines,baseline=plt.stem(n,x,label="x(n)")
plt.setp(stemlines,'linewidth',3)
#plt.show()
markerline,stemlines,baseline=plt.stem(n,x,label="x(n)")
plt.setp(stemlines,'linewidth',3)
plt.show()
```

```python
#design filter for denoising

flt = np.array([0.0,0.5,0.5,0.5])
print(flt)
```

```python
#filtering by convutition


yn=np.convolve(nssig,flt,mode='full')
```

```python
print(yn)
plt.stem(yn)
```
