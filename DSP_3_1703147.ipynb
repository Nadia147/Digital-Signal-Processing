```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#application of convulation
#1. signal denoising 

#create a signal
state = 256 #sampling rate 
t = np.arange(0,3,1/state)
points = len(t)
#print(points)
#print(t)
x=5*np.cos(2*np.pi*t)
ns=5*np.random.randn(points)
nssig=x+ns
```

```python
plt.figure(figsize=(20,10))
style.use('dark_background')

plt.plot(t,nssig,'g', label='Noisy Signal')

plt.xlabel('time in sec', fontsize=20)
plt.ylabel('Amplitude', fontsize=20)

plt.legend(fontsize=30)
plt.show()
```

```python
n=30
nssig.shape[0]
flt_sig=np.zeros(nssig.shape[0])
for i in range(0,nssig.shape[0]):
   flt_sig[i]=np.mean(nssig[i:(2*n+1)+i])

   
plt.figure(figsize=(20,10))

style.use('dark_background')

plt.plot(t, nssig, 'g-', label='NoisySignal')
plt.plot(t,flt_sig,'-r', linewidth=3,label='Filtered signal')

plt.title('Moving average signal with order=%d')

plt.legend(fontsize=15)
plt.show()
```

```python
N=100
fw=60
gtm=1000*np.arange(-N,N)/state
gfit=np.exp(-(4*np.log(2)*gtm**2)/fw**2)
gfit=gfit/np.sum(gfit)
plt.figure(figsize=(20,10))
style.use('dark_background')
plt.xlim([-100,100])
plt.plot(gtm,gfit,'m',linewidth=3,label='Gassuing filter')
plt.plot()
```

```python
sig_4_filter = np.concatenate((np.zeros(N),nssig,np.zeros(N)), axis=0)

k=len(nssig)

timeindex=np.concatenate((np.arange(-N,0),np.arange(0,k),np.arange(k,k+N)) )

time_4_filter= timeindex/state
plt.figure(figsize= (20,10))
plt.plot(time_4_filter)
```

```python
plt.figure(figsize=(20,10))
style.use('dark_background')
print(len(sig_4_filter))
plt.plot(time_4_filter,sig_4_filter,label='0 padded')
plt.show()
```

```python
plt.figure(figsize=(20,10))
style.use('dark_background')
Gfilt_sig=np.zeros(sig_4_filter.shape[0])
for i in range(0,nssig.shape[0]):
  Gfilt_sig[i]=np.sum(sig_4_filter[i:2*N+i]+gfit)
plt.plot(time_4_filter,Gfilt_sig)
plt.show()
```
