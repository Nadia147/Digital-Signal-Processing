```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style

t  = np.linspace(0,10,100)  
f = 50;              
A = 1;              
theta = np.pi/2;      


Compx_wave = A * np.exp(-1j* (2*np.pi * f * t + theta) )*np.exp(.6*t+theta) ;



plt.figure(figsize = (16,10))
style.use('dark_background')
plt.plot(t,np.real(Compx_wave),'b-',linewidth = 3, label='real part')

plt.xlim([-0.1,10])
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Complex sine wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style

t  = np.linspace(0,10,100)  
f = 50;              
A = 1;              
theta = np.pi/2;      


Compx_wave = A * np.exp(-1j* (2*np.pi * f * t + theta) )*np.exp(.6*t+theta) ;



plt.figure(figsize = (16,10))
style.use('dark_background')

plt.plot(t,np.imag(Compx_wave),'m-',linewidth = 3, label='imaginary part')
plt.xlim([-0.1,10])
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Complex sine wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```
